## Data loading

### Subtask:
Load the data from the Excel file.


**Reasoning**:
Load the Excel file into a pandas DataFrame.



```python
import pandas as pd

df = pd.read_excel("tong_san_pham_tren_dia_ban_theo_gia_hien_hanh_phan.xls")
display(df.head())
```

## Data exploration

### Subtask:
Explore the loaded DataFrame to understand its structure and identify relevant columns.


**Reasoning**:
To understand the DataFrame's structure, identify relevant columns, and check for missing values, we need to inspect the DataFrame's first few rows, last few rows, column names, data types, and missing value counts.



```python
display(df.head())
display(df.tail())
print(df.columns)
print(df.dtypes)
print(df.isnull().sum())
display(df.describe())
```

## Data wrangling

### Subtask:
Prepare the data for analysis by selecting relevant columns, handling missing values, and aggregating total product value by year.


**Reasoning**:
Create a new DataFrame `df_analysis` containing only the "Phân loại" and "Total" columns, rename the "Phân loại" column to "Year", and convert the "Year" column to numeric values.



```python
df_analysis = df[['Phân loại', 'Total']].copy()
df_analysis = df_analysis.rename(columns={'Phân loại': 'Year'})
df_analysis['Year'] = df_analysis['Year'].str.replace('Năm ', '').astype(int)
display(df_analysis)
```

**Reasoning**:
The error is caused by the row 'Sơ bộ  2023', which cannot be converted to an integer. We can handle this by filtering out this row before converting to integer.



```python
df_analysis = df[['Phân loại', 'Total']].copy()
df_analysis = df_analysis.rename(columns={'Phân loại': 'Year'})
df_analysis = df_analysis[~df_analysis['Year'].str.contains('Sơ bộ')]
df_analysis['Year'] = df_analysis['Year'].str.replace('Năm ', '').astype(int)
display(df_analysis)
```

## Data analysis

### Subtask:
Calculate the year-over-year growth rate of the total product value.


**Reasoning**:
Sort the `df_analysis` DataFrame by 'Year' and calculate the year-over-year growth rate of the 'Total' column using `pct_change()`.



```python
df_analysis = df_analysis.sort_values('Year')
df_analysis['Growth Rate'] = df_analysis['Total'].pct_change() * 100
display(df_analysis)
```

## Data visualization

### Subtask:
Create two charts: a bar chart visualizing the total product value for each year and a line chart visualizing the year-over-year growth rate.


**Reasoning**:
Create the bar chart and line chart based on the `df_analysis` DataFrame as instructed.



```python
import matplotlib.pyplot as plt

# Bar chart for total product value by year
plt.figure(figsize=(10, 5))
plt.bar(df_analysis['Year'], df_analysis['Total'])
plt.title('Total Product Value by Year')
plt.xlabel('Year')
plt.ylabel('Total Product Value')
plt.show()

# Line chart for year-over-year growth rate
plt.figure(figsize=(10, 5))
plt.plot(df_analysis['Year'], df_analysis['Growth Rate'])
plt.title('Year-over-Year Growth Rate of Total Product Value')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.show()
```
